## <font color="green">Using dictionary</h3></font>

```python
import pandas as pd
weather_data = {
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'windspeed': [6,7,2],
    'event': ['Rain', 'Sunny', 'Snow']
}
df = pd.DataFrame(weather_data)
df
```

## <font color="green">Using tuples list</h3></font>

```python
weather_data = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]
df = pd.DataFrame(data=weather_data, columns=['day','temperature','windspeed','event'])
df
```

## <font color="green">Using list of dictionaries</h3></font>

```python
weather_data = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},

]
df = pd.DataFrame(weather_data)
df
```

```python
import pandas as pd

data = {
    'Name': ['Anjali', 'Tushar', 'Chirag', 'Dev', 'Parin', 'Bhargav', 'Uvesh'],
    'Age': [24, 27, 22, 32, 29, 24, 26],
    'Department': ['HR', 'Engineering', 'Marketing', 'HR', 'Marketing', 'Engineering', 'HR'],
    'Salary': [50000, 60000, 45000, 52000, 47000, 61000, 51000],
    'Experience': [2, 4, 1, 6, 3, 5, 2]
}

df = pd.DataFrame(data)
```

```python
df.sort_values('Salary', ascending=False)
```

```python
df.sort_values(['Department','Salary']) #3 sort by department and then salary
```

```python
df.sort_values('Name', key=lambda x: x.str.len()) #(5) sort by name length
```

```python
df.sort_values(['Department','Salary'], ascending=[True, False])
```
